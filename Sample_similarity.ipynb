```python
import pandas as pd
import numpy as np
import os
import umap
import json

import networkx as nx
from gensim.models import KeyedVectors
from nodevectors import Node2Vec, GGVec

import plotly.express as px
```

```python
model = KeyedVectors.load_word2vec_format("wheel_model.bin")
model2 = KeyedVectors.load_word2vec_format("wheel_model_v2.bin")
```

```python
def get_recommendations(selections_dict, model, user_lookup, display_n=25):
    results = []
    for node_id, value in model.most_similar_cosmul(**selections_dict, topn=display_n):

        node_parts = node_id.split(':')
        if len(node_parts) != 2:
            node_type = node_parts[0]
            node_key = ' '.join(node_parts[1:])
        else:
            node_type, node_key = node_parts

        if node_type == 'Account':
            try:
                results.append((node_type, user_lookup[node_key]['Name'], value))
            except:
                print(node_key)
        elif node_key == 'True':
            pass
        else:
            results.append((node_type, node_id, value))
    df = pd.DataFrame(results, columns=['EntityType', 'Entity', 'Score'])
    return df
```

```python
with open('user_lookup.json', 'r') as f:
    user_lookup = json.load(f)
```

```python
##model with only categorical data
get_recommendations({"positive":["Account:155"]}, model, user_lookup, display_n=15)
```

```python
##model with only categorical data and numerical
get_recommendations({"positive":["City:Lisbon"]}, model2, user_lookup, display_n=45)
```
