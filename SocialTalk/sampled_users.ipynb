```python
import pandas as pd
```

```python
users = pd.read_csv('./clean/users.csv', low_memory=False).rename({'Id': 'index'}, axis=1)
user_lookup = pd.read_csv('accounts.csv', on_bad_lines='skip', low_memory=False).set_index('Id')[['Name', 'Followers', 'Estimated reach']].to_dict('index')

account_metadata = pd.DataFrame(users['index'].map(user_lookup).values.tolist(), index=users['index'])
```

```python
x = users.set_index('index').merge(account_metadata, on='index', how='right').drop(['Followers_y', 'Estimated reach_y'], axis=1)
```

```python
x.reset_index()[['Name', 'index']].rename({'index': 'Id'}, axis=1).to_csv('AvailableUsers.csv', index=None)
```

```python
x[['Name', 'Followers_x', 'Estimated reach_x']].rename({'Followers_x': 'Followers', 'Estimated reach_x': 'Estimated reach'}, axis=1).to_json('user_lookup.json', orient='index')
```
