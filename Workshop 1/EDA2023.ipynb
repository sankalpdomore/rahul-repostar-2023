```python
print('Shape',users.shape)
def EDA(df):
    a = pd.DataFrame()
    a['Features'] = df.columns
    a['Unique'] = [df[i].nunique() for i in df.columns]
    a['nan_values'] = [df[i].isna().sum() for i in df.columns]
    a['Type'] = [df[i].dtype for i in df.columns]
    return a
EDA(users)
```
